```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
#import numpy as np
from som.som import som
import matplotlib.pyplot as plt

from som.utils import somutils
```

Lectura de datos y selección de columnas

```python

df = pd.read_csv('../data/dataIrelandPopulationSOM.csv')
df.head()
columns=['avr_age','avr_education_level','avr_num_cars','unemployment_percent']
X = df[columns].to_numpy()
print('X.shape',X.shape)
```

```python

#scaler = StandardScaler()
#X_std = scaler.fit_transform(X)
#X_std.shape
```

Creación  y entrenamiento del SOM

```python
nrows = 4
ncols = 4
#dimensions = X_std.shape[1]
dimensions = X.shape[1]
vicinity = "rectangular"
# vicinity = "hexagonal"
smm = som(nrows=nrows,
          ncols=ncols,
          dimension=dimensions,
          vicinity=vicinity)
smm.train_SOM(train_data=X,
              epochs=5)
```

```python
smm.mat_count.sum()
```

Gráfica: número de datos de entrenamiento asociados a cada neurona

```python
somutils.plot_neurPointsCount(smm,figsize=(7,7))
```

Valores por neurona


```python
somutils.plot_valuesMap_2(pSom=smm, labels=columns, figsize=(9,7))
```

Usando un SOM con neuronas hexagonales

```python
nrows = 5
ncols = 6
#dimensions = X_std.shape[1]
dimensions = X.shape[1]
# vicinity = "rectangular"
vicinity = "hexagonal"
smm = som(nrows=nrows,
          ncols=ncols,
          dimension=dimensions,
          vicinity=vicinity)
smm.train_SOM(train_data=X,
              epochs=5)
```

```python
somutils.plot_neurPointsCount(smm,figsize=(7,7))
```

```python
somutils.plot_valuesMap_2(pSom=smm, labels=columns, normalize=False,figsize=(9,7))
```
